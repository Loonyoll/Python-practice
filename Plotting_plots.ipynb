```python
# Exercise 6
import numpy as np
import matplotlib.pyplot as plt

# 0 < b < 1, a - random odd number
a = int(input("Press odd number: "))
b = float(input("Press (0,1): "))
# Create set x
x = np.arange(-0.5, 0.5, 0.01)
# Create our sum
def W_x(x):
    Sum = 0.0
    for i in range(100):
        Sum += b**i * np.cos(a**i * np.pi * x)
    # Getting all values ​​of a function in the interval x
    return Sum
plt.plot(x, W_x(x), color = 'red')
plt.title(r'$\sum_{{0}}^{{n}} {}^n * \cos({}^n * pi *x)$'.format(b, a))
plt.xlabel(r'$X$')
plt.ylabel(r'$f(x)$')
plt.grid(True)
plt.show()
```
